7.  What is the difference between “==” and “is” operators in Python

-> "== operator is used to compare values, whereas is operator is used to check if two variables refer to the same object in memory."

```python
#example of 7

a = [1, 2, 3]
b = [1, 2, 3]

print(a == b)  # True (because their values are the same)
print(a is b)  # False (because they are different objects in memory)
```

9. What is type casting in Python

-> Type casting is a process of changing the data type of a value / object

-> Python provides two types of type casting:

a. Implicit Type Casting (Automatic conversion by Python): Python automatically converts a smaller data type into a larger data type to avoid data loss.

b. Explicit Type Casting (Manually converting using functions) : When automatic conversion is not possible, we use explicit type casting.



10. What is the difference between implicit and explicit type casting

-> Implicit Type Casting (Automatic Conversion)

a. Python automatically converts smaller data types to larger ones when it's safe.

b. No data loss occurs.

c. Requires no manual intervention

-> Explicit Type Casting (Manual Conversion)

a. The programmer manually converts one data type to another using built-in functions.

b. Possible data loss (e.g., float to int removes decimal part).

c. Requires explicit function calls like int(), float(), str(), etc.


```python
#implicit conversion

x = 5   # int
y = 2.5 # float
result = x + y  # int automatically converted to float
print(result)  # Output: 7.5
```

```python
#explicit conversion

x = "100"  # str
y = int(x)  # Convert str to int
print(y * 2)  # Output: 200
```

13. What is the difference between for and while loops  

-> **for Loop** is used when the number of iterations is known in advance. It iterates over a sequence (e.g., list, tuple, string, range). It is more structured and commonly used for iterating through data.

-> **while Loop** is used when the number of iterations is unknown and depends on a condition. It runs until the condition becomes False. It can result in infinite loops if the condition is not properly handled

14. Describe a scenario where a while loop is more suitable than a for loop.

-> Imagine a program that keeps asking a user for a password until they enter the correct one. In this case, we don't know how many attempts the user will take, so a while loop is the best choice.

```python
#example code

correct_password = "Python123"
attempts = 0
max_attempts = 3

while attempts < max_attempts:
    user_input = input("Enter your password: ")

    if user_input == correct_password:
        print("Access Granted!")
        break  # Exit loop if the password is correct
    else:
        attempts += 1
        print(f"Incorrect password. {max_attempts - attempts} attempts left.")

if attempts == max_attempts:
    print("Too many failed attempts. Access Denied.")
```

```python
# 6.
def demonstrate_mutable_immutable(immutable_var, mutable_var):
    print("\nBefore modification:")
    print(f"Immutable variable (int): {immutable_var}, ID: {id(immutable_var)}")
    print(f"Mutable variable (list): {mutable_var}, ID: {id(mutable_var)}")

    # Trying to modify immutable variable (integer)
    immutable_var += 10  # This creates a new integer object
    print("\nAfter modifying immutable variable:")
    print(f"Immutable variable (int): {immutable_var}, ID: {id(immutable_var)}")  # New ID

    # Trying to modify mutable variable (list)
    mutable_var.append(4)  # This modifies the existing list
    print("\nAfter modifying mutable variable:")
    print(f"Mutable variable (list): {mutable_var}, ID: {id(mutable_var)}")  # Same ID

# Initialize variables
immutable_value = 5
mutable_value = [1, 2, 3]

# Call function
demonstrate_mutable_immutable(immutable_value, mutable_value)

print("\nOutside the function after calling:")
print(f"Immutable variable (int): {immutable_value}, ID: {id(immutable_value)}")  # Unchanged
print(f"Mutable variable (list): {mutable_value}, ID: {id(mutable_value)}")  # Modified
```

```python
# 7. Write a program that performs basic arithmetic operations on two user-input numbers.

def add(x, y):
    return x + y

def subtract(x, y):
    return x - y

def multiply(x, y):
    return x * y

def divide(x, y):
    if y == 0:
        return "Error! Division by zero is not allowed."
    return x / y

def main():
    print("Basic Arithmetic Operations")
    try:
        num1 = float(input("Enter first number: "))
        num2 = float(input("Enter second number: "))

        print("Select operation:")
        print("1. Addition")
        print("2. Subtraction")
        print("3. Multiplication")
        print("4. Division")

        choice = input("Enter choice (1/2/3/4): ")

        if choice == '1':
            print(f"Result: {add(num1, num2)}")
        elif choice == '2':
            print(f"Result: {subtract(num1, num2)}")
        elif choice == '3':
            print(f"Result: {multiply(num1, num2)}")
        elif choice == '4':
            print(f"Result: {divide(num1, num2)}")
        else:
            print("Invalid choice! Please select a valid operation.")
    except ValueError:
        print("Invalid input! Please enter numeric values only.")

if __name__ == "__main__":
    main()
```

```python
# 9. Write a Python program to convert user input from string to integer, float, and boolean types

def convert_input(user_input):
    try:
        int_value = int(user_input)
        print(f"Integer value: {int_value}")
    except ValueError:
        print("Cannot convert to integer.")

    try:
        float_value = float(user_input)
        print(f"Float value: {float_value}")
    except ValueError:
        print("Cannot convert to float.")

    bool_value = user_input.strip().lower()
    if bool_value in ['true', '1', 'yes', 'on']:
        print("Boolean value: True")
    elif bool_value in ['false', '0', 'no', 'off']:
        print("Boolean value: False")
    else:
        print("Cannot convert to boolean.")

if __name__ == "__main__":
    user_input = input("Enter a value: ")
    convert_input(user_input)
```

```python
#13. Write a Python program to find the sum of all even numbers between 1 and 50.

sum_even = 0

# Initialize counter
num = 2  # Start from the first even number

# Use a while loop to add even numbers up to 50
while num <= 50:
    sum_even += num  # Add the even number to sum
    num += 2  # Move to the next even number

# Print the result
print(f"Sum of all even numbers between 1 and 50 is: {sum_even}")
```

```python
#14. Write a program to reverse a string using a while loop.

# Get input from the user
text = input("Enter a string: ")

# Initialize an empty string to store the reversed result
reversed_text = ""

# Initialize index to the last character of the string
i = len(text) - 1

# Reverse the string using a while loop
while i >= 0:
    reversed_text += text[i]  # Append the character at index i
    i -= 1  # Move to the previous character

# Print the reversed string
print(f"Reversed string: {reversed_text}")
```

```python
# 15. Write a Python program to calculate the factorial of a number provided by the user using a while loop

# Get input from the user
num = int(input("Enter a number: "))

# Initialize factorial variable
factorial = 1

# Initialize counter
i = num

# Calculate factorial using while loop
while i > 0:
    factorial *= i  # Multiply current number with factorial
    i -= 1  # Decrease counter

# Print the result
print(f"Factorial of {num} is {factorial}")
```
